Creo las bases de datos

```python
import pandas as pd
import numpy as np
from numpy import random
import matplotlib.pyplot as plt

data = pd.read_csv(r'meteorite-landings.csv')
data
```

```python
meteoritos_h4 = data[data.recclass.str.contains('H4')]
meteoritos_h4 = meteoritos_h4[~meteoritos_h4.recclass.str.contains('E')]
meteoritos_h4 = meteoritos_h4[~meteoritos_h4.recclass.str.contains('5')]
```

```python
meteoritos_h4
```

```python
meteoritos_h5 = data[data.recclass.str.contains('H5')]
meteoritos_h5 = meteoritos_h5[~meteoritos_h5.recclass.str.contains('E')]
#meteoritos_h5 = meteoritos_h5[~meteoritos_h5.recclass.str.contains('5')]
```

```python
meteoritos_h5
```

Limpio los datos de Null, Nan, ceros y años mayores a 2021

```python
meteoritos_h4 = meteoritos_h4.dropna()
meteoritos_h4 = meteoritos_h4[(meteoritos_h4.year < 2022)]
meteoritos_h4 =meteoritos_h4[meteoritos_h4['mass'] !=0]

print(meteoritos_h4.isnull().sum())
print(meteoritos_h4['year'].max())
print(meteoritos_h4['mass'].min())
```

```python
meteoritos_h5 = meteoritos_h5.dropna()
meteoritos_h5 = meteoritos_h5[(meteoritos_h5.year < 2022)]
meteoritos_h5 =meteoritos_h5[meteoritos_h5['mass'] !=0]

print(meteoritos_h5.isnull().sum())
print(meteoritos_h5['year'].max())
print(meteoritos_h5['mass'].min())
```

Cuento los meteoritos entre 1960 y 1990

```python
year_rangeH4 = ((meteoritos_h4.year >= 1960) & (meteoritos_h4.year <= 1990))
year_rangeH5 = ((meteoritos_h5.year >= 1960) & (meteoritos_h5.year <= 1990))

print(len(meteoritos_h4[year_rangeH4]),len(meteoritos_h5[year_rangeH5]))
```

Concateno y grafico

```python
meteoritos_h4h5 = pd.concat([meteoritos_h4, meteoritos_h5]) 
meteoritos_h4h5
```

```python
data_yearh4 = meteoritos_h4.groupby('year').agg({'year':'max','recclass':'count'})
data_yearh5 = meteoritos_h5.groupby('year').agg({'year':'max','recclass':'count'})
```

```python

plt.plot(data_yearh4['year'], data_yearh4['recclass'], label = 'H4')
plt.plot(data_yearh5['year'], data_yearh5['recclass'], label = 'H5')
plt.title("Number of meteorites by class and by year")
plt.xlabel("Year")
plt.ylabel("Number of meteorites")
plt.legend();
```

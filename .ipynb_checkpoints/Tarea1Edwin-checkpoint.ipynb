```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('oec.csv')
pd.set_option('display.max_columns', None)
# pd.reset_option('max_columns')
data.head()
```

```python
methods = data.groupby('DiscoveryMethod').sum()
methods.head()
```

```python
Massbymethods = methods.agg('PlanetaryMassJpt')
Massbymethods.head()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
methods = ['RV', 'imaging', 'microlensing', 'timing', 'transit']
ax.bar(methods,Massbymethods)
plt.show()
```

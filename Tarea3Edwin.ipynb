9. Graficaré los planetas descubiertos por año y por Radial Velocity:

```python
radial = datos[datos.method.str.contains('Radial')]
data_radial = radial.groupby('year').agg({'year':'max','number':'count'})
plt.plot(data_radial['year'], data_radial['number'])
#plt.plot(data_yearh5['year'], data_yearh5['recclass'], label = 'H5')
plt.title("Number of Exoplanet discoveries by Radial Velocity and year")
plt.xlabel("Year")
plt.ylabel("Number")
#plt.legend();
```

10 y 11: Le genero las dos anotaciones

```python
fig, ax = plt.subplots(figsize=(12, 4))
plt.plot(data_radial['year'], data_radial['number'])
plt.title("Number of Exoplanet discoveries by Radial Velocity and year")
plt.xlabel("Year")
plt.ylabel("Number")
#ax.text(1989, 1, ". Data: (1, 5)", transform=ax.transData)
ax.text(1988, 5, "First Discovery", color='red')
#plt.legend()
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
plt.plot(data_radial['year'], data_radial['number'])
plt.title("Number of Exoplanet discoveries by Radial Velocity and year")
plt.xlabel("Year")
plt.ylabel("Number")
#ax.text(1989, 1, ". Data: (1, 5)", transform=ax.transData)
ax.text(1988, 5, "First Discovery", color='red')
ax.annotate('Maximum discoveries in a single year', xy=(2011, 94), xytext=(2000, 70),
            arrowprops=dict(facecolor='black', shrink=0.05))
#plt.legend()
```
